```python
import cv2
import numpy as np
from PIL import Image

# 투명한 PNG 이미지 로드 및 크기 조정
overlay = Image.open("free-icon-face-recognition-5061016.png").convert("RGBA")
overlay_resized = overlay.resize((360, 360))

# 웹캠 열기
cap = cv2.VideoCapture("rtsp://darkice:sys3275423@192.168.0.166:554/stream1")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_resized = cv2.resize(frame, (640, 360), interpolation=cv2.INTER_AREA)
    frame_resized = frame_resized[0:360, 140:500].copy()

    # OpenCV 이미지를 PIL 이미지로 변환
    frame_pil = Image.fromarray(cv2.cvtColor(frame_resized, cv2.COLOR_BGR2RGB)).convert("RGBA")

    # 웹캠 중앙 좌표 계산 (왼쪽 상단 기준)
    x_offset = 0 # X 좌표
    y_offset = 0  # Y 좌표 (위쪽 정렬)
    
    # 빈 배경 생성 및 합성
    frame_pil.paste(overlay_resized, (x_offset, y_offset), overlay_resized)

    # 다시 OpenCV 형식으로 변환하여 출력
    combined_cv = cv2.cvtColor(np.array(frame_pil), cv2.COLOR_RGBA2BGR)
    cv2.imshow("Overlay", combined_cv)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
